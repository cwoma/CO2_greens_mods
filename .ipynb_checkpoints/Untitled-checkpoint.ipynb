## Create a compounding concentration profile

```python
init_co2 = 2.21718909e+15 #estimated from initial co2 mass above from control ***need to fix*** #284.3 #ppm https://gmd.copernicus.org/articles/10/2057/2017/
pct_increase = 0.01
years_increase = 150
co2_arr = utils.compound_mult(init_co2, years_increase, pct_increase)
```

## Fit an emissions profile convolution with the concentration above

```python
#test_conv = utils.convolve_single_lev_mean(G, np.linspace(1,90, 151), 1)
test_conv = signal.convolve(G , np.linspace(1,70, 151), mode = 'full')
#utils.compound_mult(1, 150, .035)
```

```python
plt.plot(init_co2 + test_conv, label = 'convolution', marker = '*')
plt.plot(co2_arr, label = 'model')
plt.legend()
plt.xlim(0,150)
plt.ylim(0,1e16)
```
